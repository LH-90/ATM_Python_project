```python
pin_user = 1062
balance_user = 1879

def main_menu():
    print("Hello, what would you like to do?")
    print("1: Displaying the bank account balance")
    print("2: Withdrawing money")
    print("3: Depositing money")
    print("4: Exiting from the ATM")

def balance():
  print("The current balance is: " + str(balance_user) + "£.")
  return balance_user

def main_menu_option():


  while True:
     user_option = int(input("Choose an option: (1-4) "))
     if user_option == 1:
        balance()

     elif user_option == 2:
        withdraw()

     elif user_option == 3:
        deposit()

     elif user_option == 4:
        print("Thank you for using the ATM")
        break

     else:
        print("Invalid. Please enter a valid number.")
        main_menu_option()

     user_input = input("Would you like to perform another operation? (yes/no): ")

     if user_input.lower() == "yes":
         continue

     else:
         print("Thank you for using the ATM")
         break


def atm():
    attempts = 0

    while attempts < 3:
        input_pin = int(input("Enter your PIN: "))


        if input_pin == pin_user:
            print("PIN verification successful.")
            main_menu()
            main_menu_option()
            break


        else:
            attempts += 1
            if attempts == 1 or attempts == 2:
              print("Invalid PIN. Please try again.")

    else:
        print("Your account has been locked.")





def withdraw():
  global balance_user
  print("How much money would you like to withdraw?")
  print("1: 10£")
  print("2: 20£")
  print("3: 30£")
  print("4: 50£")
  print("5: 80£")
  print("6: 100£")
  print("7: Other amount")

  input_amount = int(input("Enter an amount: (1-7)"))
  numbers = [1, 2, 3, 4, 5, 6]
  amounts = [10, 20, 30, 50, 80, 100]
  numbers_amounts = {num: amount for num, amount in zip(numbers, amounts)}

  if input_amount in numbers :

      if numbers_amounts[input_amount] < balance_user:
       balance_user -= numbers_amounts[input_amount]
       print(numbers_amounts[input_amount], "£ has been withdrawn. The new balance is ", balance_user, "£.")


      else:
       print("This amount cannot be withdrawn. You only have", balance_user, "£ on your account.")

  while input_amount == 7:
       input_amount2 = int(input("Enter another amount: "))

       if input_amount2 < balance_user:

         if input_amount2 % 10 == 0:
            balance_user -= input_amount2
            print(str(input_amount2) + "£ has been withdrawn. The new balance is " + str(balance_user) + "£.")

         else:
            print("This amount cannot be withdrawn. Try another one.")

       else:
         print("This amount cannot be withdrawn. You have less than", input_amount2, "on your account.")
       break

def deposit():
  global balance_user
  print("How much money would you like to deposit?")
  input_deposit = int(input("Enter an amount: "))
  balance_user += input_deposit
  print("Thank you, your new balance is " + str(balance_user) + "£.")
  return balance_user
atm()
```
